```python
import numpy as np
import pandas as pd
from datetime import datetime

from matplotlib import pyplot as plt
import seaborn as sns
```

```python
annual_fares = pd.read_csv('annual_fares_1995_2022_Q2of2022.csv')
```

```python
annual_fares
```

```python
melted_fares = pd.melt(annual_fares, id_vars=['year'], value_vars=['inf_adj_avg_fare', 'avg_fare'], var_name='adj_or_not', value_name='fare')
```

```python
melted_fares
```

```python
year_list = []
for i in range(1995, 2023):
    year_list.append(i)
```

```python
plt.figure(figsize=(6, 3), dpi=300)

sns.set(font_scale = 0.6)

ax = sns.lineplot(data=melted_fares, x='year', y='fare', hue='adj_or_not', palette=["#80B1D3","#FDB462"], legend=False,)

plt.xticks(year_list, rotation=60)

plt.legend(title='Inflation Adjusted Y/N', loc='upper right', labels=['Inflation Adjusted', 'Original'])
plt.title('Annual Average Flight Fares (1995 to 2022)')
plt.xlabel('Year')
plt.ylabel('Annual Average Fare in dollars')

plt.show()
```
