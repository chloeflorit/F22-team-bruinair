```python
grouped = flights.groupby("year")
year_2018 = grouped.get_group(2018)
year_2019 = grouped.get_group(2019)
year_2020 = grouped.get_group(2020)
year_2021 = grouped.get_group(2021)
year_2022 = grouped.get_group(2022)
```

```python
summary = flights.groupby(["year", "quarter"])[["market_fare"]].mean()
summary
```

```python
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
colors = ["#FB8072", "#80B1D3","#BC80BD","#B3DE69"]
sb.set_palette(sb.color_palette(colors))
# perform groupby
summary = summary.reset_index()
  
# plot barplot
g = sb.barplot(x="year",
           y="market_fare",
           hue="quarter",
           data=summary)

g.set_title("Average Flight Price Per Quarter from 2018 to 2022")
g.set_ylabel("Average flight price")
g.set_xlabel("Year")
g.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0, title = "Quarter")
```
